```python
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
x = dataset.iloc[:,3:5].values
```

```python
x
```

```python
from sklearn.cluster import AffinityPropagation
cluter = AffinityPropagation(random_state=42)
y_Cluter = cluter.fit_predict(x)
```

```python
superviced=dataset
```

```python
superviced
```

```python
superviced['Cluter_group'] = y_Cluter
```

```python
superviced
```

```python
y_Cluter
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lmplot(data=superviced,x=superviced.columns[3],y=superviced.columns[4],hue=superviced.columns[5],legend=True, fit_reg=False, )
```
